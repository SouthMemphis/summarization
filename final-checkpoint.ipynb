### Transformers have three main architectures: Encoder, Decoder, and Encoder-Decoder. The Encoder component of the Transformer architecture is responsible for processing the input sequence, often a piece of text, and extracting important features while creating a rich internal representation of the information.Â¶Decoder component takes the encoded representation from the Encoder and generates the output sequence, which is typically the summarized version of the input text. Encoder-Decoder architecture in Transformers is a powerful framework that excels in the task of summarization due to its ability to understand the context of the input (handled by the Encoder) and generate appropriate outputs (handled by the Decoder).

### And yet, when it comes to learning the BART model (or any other) for summarization tasks, it becomes obvious that this requires a lot of computing power. Here where distillation technique comes.

### Knowledge distillation is used to compress a complex and large neural network into a smaller and simpler one, while still retaining the accuracy and performance of the resultant model. This process involves training a smaller neural network to mimic the behavior of a larger and more complex "teacher" network by learning from its predictions or internal representations.

### Below you can see the general concept of this idea

### Ok, but what about summarization of long-text documents? Summarizing large texts requires significant computational resources, which are not always available. To address this challenge, models like Longformer-encoder-decoder have been developed. These models modify the attention mechanism to minimize computational requirements during fine-tuning and inference. By optimizing attention, they aim to reduce resource consumption while maintaining summarization quality. This enables more efficient processing of long texts, making it feasible to summarize them even with limited computational resources.


# Let's see, how it can be achieved 

### Instead of using the usual attention mechanism, LED uses a whole combination of a slightly different approach to attention

 - ### Sliding Window Attention
 - ### Dilated Sliding Window
 - ### Global + Sliding Window

### Longformer uses sliding window attention to build local context, applying the attention mechanism on neighboring tokens. 

```python
Image(filename = "C:/Users/yaroslav/Desktop/fig-2.png", width=2000, height=2000)
```

### Multiple stacked layers of sliding window attention results in a large receptive ï¬eld. The idea is similar to convolutional neural networks (CNNs) where top layers have access to all input locations, building global semantic representations. Through  *l* layers, the receptive field size of the topmost layer is  *l* * *w*, where *w* is the window size (here, assuming the window size  is fixed through all layers). The computational complexity is *O(n * w)*, which scales linearly with the input sequence length.
